```python
# Import additional libraries. 

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing our data

df = pd.read_csv('ds_salaries.csv')
df.head(5)
```

```python
# Lets lis our columns
df.columns
```

```python
df.drop(columns = 'id', inplace=True)
df.columns
```

```python
residence_grp = df.groupby('employee_residence')
```

```python
us_employees = residence_grp.get_group('US')
us_employees.head(3)
```

```python
experience_level = us_employees['experience_level'].unique()
experience_level
```

```python
us_employees.query('experience_level == "Entery Level"')['salary'].median()
```

```python
saleries = []
exp_level = []

for value in experience_level:
    saleries.append(int(us_employees.query(f'experience_level == "{value}"')['salary'].median()))
    exp_level.append(f'{value}')
    
    
```

```python
# Plotting our data
plt.bar(exp_level, saleries, color = 'skyblue')

#labeling our axes
plt.xlabel('Experience level')
plt.ylabel('Saleries')
plt.title('Saleries and Experience Level')

# Rotate our x-axis labels for improved reading
plt.xticks(rotation=45)

plt.show()
```

```python
# Plotting our data
barplot = plt.bar(exp_level, saleries, color = 'skyblue')
plt.bar_label(barplot, labels=saleries, label_type='center')

#labeling our axes
plt.xlabel('Experience level')
plt.ylabel('Saleries in USD')
plt.title('Saleries and Experience Level')

# Rotate our x-axis labels for improved reading
plt.xticks(rotation=45)

plt.show()
```

```python
# Plotting our data
barplot = plt.bar(exp_level, saleries, color = 'skyblue')
plt.bar_label(barplot, labels=saleries, label_type='edge')

#labeling our axes
plt.xlabel('Experience level')
plt.ylabel('Saleries in USD')
plt.title('Saleries and Experience Level')

# Rotate our x-axis labels for improved reading
plt.xticks(rotation=45)

plt.show()
```
